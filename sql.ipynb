```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_buses_1=pd.read_csv("df_buses_1.csv")
df_buses_2=pd.read_csv("df_buses_2.csv")
df_buses_3=pd.read_csv("df_buses_3.csv")
df_buses_4=pd.read_csv("df_buses_4.csv")
df_buses_5=pd.read_csv("df_buses_R.csv")
df_buses_6=pd.read_csv("df_buses_SB.csv")
df_buses_7=pd.read_csv("df_buses_H.csv")
df_buses_8=pd.read_csv("df_buses_AS.csv")
df_buses_9=pd.read_csv("df_buses_9.csv")
df_buses_10=pd.read_csv("df_buses_10.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

DATA CLEANING

```python
#get about the data
Final_df.info()
```

```python
#convert ratings into numeric

# Convert the Ratings column to string type to safely apply string operation
Final_df["Ratings"] = Final_df["Ratings"].astype(str)

# Remove the word "New"
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")

# Strip any leading/trailing whitespace
Final_df["Ratings"]=Final_df["Ratings"].str.strip()

# Extract the first character (assuming this is the numeric part you want)
Final_df["Ratings"]=Final_df["Ratings"].str.strip().str[0]

# Convert the result to numeric, setting errors='coerce' to handle non-numeric values
Final_df["Ratings"]=pd.to_numeric(Final_df["Ratings"], errors='coerce')

# Fill NaN values with 0
Final_df.fillna({"Ratings": 0}, inplace=True)
```

```python
#covert prices to numeric
# Remove the "INR" string from all entries in the "Price" column
Final_df["Price"]=Final_df["Price"].str.replace("INR","")

# Convert the cleaned "Price" column entries to floating-point numbers
Final_df["Price"]=Final_df["Price"].astype(float)


# Replace any NaN values in the "Price" column with 0
Final_df["Price"].fillna(0)
```

```python
Final_df.info()
```

```python
#replace nan value
Final_df =  Final_df.replace({np.nan : None})
```

```python
# Filter the DataFrame to include only rows where the "Price" is less than or equal to 7000
Final_df =  Final_df[Final_df["Price"] <= 7000]
```
